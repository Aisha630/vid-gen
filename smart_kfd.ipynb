```python
path_to_video = "/root/VideoReconstruction/our_results/lambshery_interpolated.mp4"
outout_color_corrected = "/root/VideoReconstruction/our_results/lambshery_interpolated_color_corrected.mp4"

import cv2
from  diffusers.utils import export_to_video

import PIL
import IPython.display as display
# Input and output video paths
input_video_path = "/root/VideoReconstruction/our_results/lambshery_interpolated.mp4"
output_video_path =  "/root/VideoReconstruction/our_results/lambshery_interpolated_color_corrected.mp4"

# Open the input video
cap = cv2.VideoCapture(input_video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4

# Create a VideoWriter object
# out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Function to normalize contrast using CLAHE
def normalize_contrast(frame):
    # Convert to LAB color space
    lab = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)

    # Apply CLAHE to the L-channel
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    l_normalized = clahe.apply(l)

    # Merge the normalized L-channel back with a and b
    lab_normalized = cv2.merge((l_normalized, a, b))

    # Convert back to BGR color space
    frame_normalized = cv2.cvtColor(lab_normalized, cv2.COLOR_LAB2BGR)

    return frame_normalized

# Process each frame
frames = []
while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break

    # Normalize the frame
    normalized_frame = normalize_contrast(frame)
    
    frames.append(normalized_frame)
    # Display the normalized frame
    # cv2.imshow('Normalized Frame', normalized_frame)
    display.display(PIL.Image.fromarray(normalized_frame))

    # Write the normalized frame to the output video
    # out.write(normalized_frame)

# Release resources\
export_to_video(frames, output_video_path, fps)
cap.release()
# out.release()

print(f"Normalized video saved to {output_video_path}")

```
